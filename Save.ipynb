```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [45,34,56,23,45]

plt.subplot(2,2,1)
plt.step(x,y)

x = [5,6,7,8,9]
y = [67,50,66,56,82]

plt.subplot(2,2,2)
plt.bar(x,y)

x = [2,4,6,8,10]
y = [57,50,60,55,20]

plt.subplot(2,2,3)
plt.stackplot(x,y)

x = [1,3,5,7,9]
y = [67,51,62,50,20]

plt.subplot(2,2,4)
plt.scatter(x,y)

plt.savefig("Demo.png",pad_inches = 0.9,bbox_inches = "tight",facecolor="gray")
plt.show()
```

```python
import matplotlib.pyplot as plt

brands = ["OnePlus","Apple","Samsung","Nokia","Redmi"]
x = [10,40,30,3,17]
c=["violet","silver","red","teal","yellow"]
ex = [0,0,0,0,0.1]
plt.pie(x,labels = brands,colors = c,explode = ex,shadow = True,autopct = '%.2f',startangle=180)

plt.savefig("brand popularity.png",facecolor="black")
plt.show()
```
