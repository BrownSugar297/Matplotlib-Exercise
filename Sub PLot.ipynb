```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [45,34,56,23,45]

plt.subplot(2,2,1)
plt.title("Age")
plt.bar(x,y)

x = [5,6,7,8,9]
y = [67,50,66,56,82]

plt.subplot(2,2,2)
plt.title("Weight")
plt.plot(x,y)

plt.suptitle("INFO")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [45,34,56,23,45]

plt.subplot(2,2,1)
plt.step(x,y)

x = [5,6,7,8,9]
y = [67,50,66,56,82]

plt.subplot(2,2,2)
plt.bar(x,y)

x = [2,4,6,8,10]
y = [57,50,60,55,20]

plt.subplot(2,2,3)
plt.stackplot(x,y)

x = [1,3,5,7,9]
y = [67,51,62,50,20]

plt.subplot(2,2,4)
plt.scatter(x,y)

plt.show()
```
