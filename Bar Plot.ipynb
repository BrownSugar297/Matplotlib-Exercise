```python
import matplotlib.pyplot as plt

x = ["Part1","Part2","Part3","Part4","Part5"]
y = [98,67,88,95,88]

colors = ["red","blue","green","yellow","gray"]
plt.bar(x,y, color =colors,edgecolor = "black")
plt.xlabel("Parts of Harry Potter",fontsize = 20)
plt.ylabel("Popularity",fontsize = 30)
plt.title("Popularity of different part of Harry Potter",fontsize = 5)
plt.show
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("expense3.xlsx")
df = pd.DataFrame(data)
print(df)
print()
grouped_by=df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)
print()

plt.bar(grouped_by.index,grouped_by.values)
plt.show()
```
