```python
import matplotlib.pyplot as plt

brands = ["OnePlus","Apple","Samsung","Nokia","Redmi"]
x = [10,40,30,3,17]
c=["violet","silver","red","teal","yellow"]
ex = [0,0,0,0,0.1]
plt.pie(x,labels = brands,colors = c,explode = ex,shadow = True,autopct = '%.2f',startangle=180)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel("expense3.xlsx")
df
```

```python
plt.pie(df["Amount"])
plt.show()
```

```python
grouped_by = df.groupby("Payment Mode")["Amount"].sum()
print(grouped_by)
```

```python
c =["yellow","red","blue"]
ex = [0,0,.1]
plt.pie(grouped_by.values,labels=grouped_by.index,colors=c,explode=ex,shadow=True,autopct="%.f")
plt.show()
```
