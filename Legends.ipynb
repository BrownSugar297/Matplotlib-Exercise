```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [45,67,33,67,12]
y1 = [41,60,13,66,13]
y2 = [10,30,55,60,19]

plt.figure(figsize = [5,3])
plt.plot(x,y,label="male")
plt.plot(x,y1,label="female")
plt.plot(x,y2,label="kids")

plt.legend(["A1","A2","A3"],loc=0)#0,1,2,3,4,5......10
plt.show()
```
