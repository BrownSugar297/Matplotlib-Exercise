```python
import matplotlib.pyplot as plt

Days = [1,2,3,4,5,6,7]

NOP1 = [5,10,30,20,35,60,80]
NOP2 = [50,60,30,75,40,90,120]
NOP3 = [8,30,50,60,70,90,100]

plt.stackplot(Days,NOP1,NOP2,NOP3)
plt.show()
```

```python
plt.stackplot(Days,NOP1,NOP2,NOP3,baseline="sym",colors=["pink","red","teal"],labels=["week1","week2","week3"])
plt.legend()
plt.show()
```
